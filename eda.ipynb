```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import matplotlib.pyplot as plt


dataset, info = tfds.load('oxford_iiit_pet:3.*.*', with_info=True)
# exploratory data analysis

# Display basic information about the dataset
print(info)

# Display the number of samples in each split
print(f"Total number of samples in the dataset: {info.splits['train'].num_examples + info.splits['test'].num_examples}")
print(f"Number of samples in the training set: {info.splits['train'].num_examples}")
print(f"Number of samples in the test set: {info.splits['test'].num_examples}")

# Display some sample images from the dataset
def display_sample_images(dataset, num_samples=5):
    plt.figure(figsize=(10, 2*num_samples))
    for i, datapoint in enumerate(dataset['train'].take(num_samples)):
        image = datapoint['image']
        plt.subplot(num_samples, 1, i+1)
        plt.imshow(image.numpy())
        plt.title(datapoint['label'].numpy())
        plt.axis('off')
    plt.show()

display_sample_images(dataset)
```
